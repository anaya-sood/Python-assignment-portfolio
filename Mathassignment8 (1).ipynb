```python
import numpy as np 

def gradient_descent(x0, y0, f, grad_f, alpha, num_iterations):
    """
    Parameters:
    x0, y0: Initial point for the descent.
    f: a function of two variables
    grad_f: the gradient of f
    alpha: Learning rate.
    num_iterations: Number of iterations to perform.

    Returns:
    (x, y): The coordinates of the final point after gradient descent.
    """
    x, y = x0, y0  # Initialize x and y with the initial point

    for i in range(num_iterations):
        grad_x, grad_y = grad_f(x, y)  # compute gradient
        x = x - alpha * grad_x  # update x
        y = y - alpha * grad_y  # update y

    return x, y
```

```python
def fun_1(x, y):
    return x**2 + y**2

def grad_f_1(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y

# Test Case 1
result_1 = gradient_descent(0.1, 0.1, fun_1, grad_f_1, alpha=0.1, num_iterations=10)
print("Result 1:", result_1)

# Test Case 2
result_2 = gradient_descent(-1, 1, fun_1, grad_f_1, alpha=0.01, num_iterations=100)
print("Result 2:", result_2)
```

```python
import numpy as np

def fun_2(x, y):
    return 1 - np.exp(-x**2 - (y - 2)**2) - 2 * np.exp(-x**2 - (y + 2)**2)

def grad_f_2(x, y):
    grad_x = 2 * x * (np.exp(-x**2 - (y - 2)**2) + 2 * np.exp(-x**2 - (y + 2)**2))
    grad_y = 2 * ((y - 2) * np.exp(-x**2 - (y - 2)**2) + 2 * (y + 2) * np.exp(-x**2 - (y + 2)**2))
    return grad_x, grad_y

# Test Case 1
result_3 = gradient_descent(0, 1, fun_2, grad_f_2, alpha=0.01, num_iterations=10000)
print("Result 3 (start at (0, 1)):", result_3)

# Test Case 2
result_4 = gradient_descent(0, -1, fun_2, grad_f_2, alpha=0.01, num_iterations=10000)
print("Result 4 (start at (0, -1)):", result_4)
```
