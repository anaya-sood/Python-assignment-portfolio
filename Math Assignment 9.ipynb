```python
import sympy as sp

x, y = sp.symbols('x y')
f = sp.exp(x)*sp.sin(y) + y**3

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

df_dx, df_dy
```

```python
g = x**2 * y + x * y**2

grad_g = [sp.diff(g, var) for var in (x, y)]
grad_g_at_point = [expr.subs({x: 1, y: -1}) for expr in grad_g]

# Magnitude of gradient vector
magnitude = sp.sqrt(sum(val**2 for val in grad_g_at_point))
grad_g, grad_g_at_point, magnitude
```

```python
h = sp.ln (x**2 + y**2)
d2h_dx2 = sp.diff (h,x,2)
d2h_dy2 = sp.diff (h,y,2)
d2h_dxdy = sp. diff (h,x,y) 
d2h_dx2, d2h_dy2, d2h_dxdy
```

```python
import numpy as np
import matplotlib.pyplot as plt

j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
k = sp.cos(x*y) + x**2 - y**2
k_func = sp.lambdify((x, y), k, 'numpy')

Z2 = k_func(X, Y)

plt.contourf(X, Y, Z2, levels=50, cmap='plasma')
plt.colorbar()
plt.title(r'Contour plot of $k(x, y) = \cos(xy) + x^2 - y^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
